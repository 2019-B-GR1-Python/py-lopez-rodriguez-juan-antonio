```python
class Main:
    
    def menu():
        print("\nSeleccione la opción deseada:\n------------PROFESORES------------\n1: Leer\n2: Nuevo\n3: Eliminar\n4: Actualizar\n5: Buscar")
        print("\n------------COLEGIOS-----------\n6: Leer\n7: Nuevo\n8: Eliminar\n9: Actualizar\n10: Buscar\n\n11: SALIR")
        
    def leer():
        try:
            path = "./ColegioProfesores.txt"
            archivo_abierto = open(path)
            contenido = archivo_abierto.readlines()
            
            for linea in contenido:
                print(linea)
            archivo_abierto.close()
            
        except Exception as error:
            print("Error")
    
    def leercolegios():
        try:
            path = "./Colegio.txt"
            archivo_abierto = open(path)
            contenido = archivo_abierto.readlines()
            
            for linea in contenido:
                print(linea)
            archivo_abierto.close()
            
        except Exception as error:
            print("Error")     
    
    
    while True:
        menu()
        
        opcion_menu = input()
        
        if(opcion_menu == "1"):
            #LEER PROFESOR
            print("")
            leer()
        
        elif(opcion_menu == "2"):
            print("")
            # NUEVO PROFESOR
            id_profesor= input("Ingrese el id del profesor-->")
            nombre= input("Ingrese el nombre del profesor-->")
            sueldo= input("Ingrese el sueldo-->")
            especialidad= input("Ingrese la especialidad-->")
            colegio= input("Ingrese el colegio-->")
            profesor= [(id_profesor, nombre,sueldo,especialidad, colegio)]
            Profesor.insertar_profesor(profesor)
            print("Profesor insertado correctamente")
        
        
        elif(opcion_menu == "3"):
            #ELIMINAR PROFESOR
            print("")
            eliminar= input("Ingrese el nombre del profesor que desea eliminar-->")
            Profesor.eliminar_profesor(eliminar)
    
        elif(opcion_menu == "4"):
            print("")
            # ACTUALIZAR PROFESOR
            modificar= input("Ingrese el nombre del profesor que desea modificar-->")
            Profesor.modificar_profesor(modificar)
            
        elif(opcion_menu == "5"):
            #BUSCAR PROFESOR
            print("")
            print("Para buscar:")
            busca= input("Ingrese el nombre del profesor que desea buscar-->")
            Profesor.buscar_profesor(busca)

        
        elif(opcion_menu == "6"):
            #LEER COLEGIO
            print("")
            leercolegios()
        
        elif(opcion_menu == "7"):
            print("")
            # NUEVO COLEGIO
            id_colegio= input("Ingrese el id del colegio: ")
            nombrecol= input("Ingrese el nombre del colegio: ")
            aforo= input("Ingrese el aforo: ")
            direccion= input("Ingrese la direccion: ")
            colegio= [(id_colegio, nombrecol,aforo,direccion)]
            Colegio.insertar_colegio(colegio)
            print("Colegio insertado correctamente")
        
        
        elif(opcion_menu == "8"):
            #ELIMINAR COLEGIO
            print("")
            eliminar= input("Ingrese el nombre del colegio que desea eliminar: ")
            Colegio.eliminar_colegio(eliminar)
    
        elif(opcion_menu == "9"):
            print("")
            # ACTUALIZAR COLEGIO
            actualizar= input("Ingrese el nombre del colegio que desea actualizar: ")
            Colegio.modificar_colegio(actualizar)
            
        elif(opcion_menu == "10"):
            #BUSCAR COLEGIO
            print("")
            print("Para buscar:")
            buscar= input("Ingrese el nombre del colegio que desea buscar: ")
            Colegio.buscar_colegio(buscar)

        
        elif(opcion_menu == "11"):
            break
        
        else:
            print ("")
            input("No ha pulsado ninguna opción correcta...\nPulse una tecla para continuar")
```
