```python
class Colegio: 
    nombre = None 
    direccion = None 
    aforo_estudiantes = None

    def __init__(self, nombre, direccion, aforo_estudiantes):   
        self.nombre = nombre
        self.direccion = direccion
        self.aforo_estudiantes = aforo_estudiantes

    def imp():
        print("si vale")
    
    def crear_colegio():
        nombre = input("Ingrese el nombre del colegio: ")
        direccion = input("Ingrese la direccion del colegio: ")
        aforo_estudiantes = input("Ingrese el aforo de estudiantes del colegio: ")
        nuevo_colegio = Colegio(nombre, direccion, aforo_estudiantes)
        print(nuevo_colegio.nombre, nuevo_colegio.direccion, nuevo_colegio.aforo_estudiantes)

        try:
            path = './colegios.txt'
            archivo_escritura_abierto = open(path,mode="a")
            archivo_escritura_abierto.writelines(
            ["\n Colegio: " + nuevo_colegio.nombre + " - Direccion: ", nuevo_colegio.direccion + " - Aforo: ", nuevo_colegio.aforo_estudiantes + " - "]
            )

        except Exception as error:
            print("Error")


    def imprime(self):
        return f"Nombre: {self.nombre}" + f"  Direccion: {self.direccion}" + f"  Aforo de estudiantes: {self.aforo_estudiantes}"


```
