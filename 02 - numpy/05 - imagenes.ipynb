```python
import numpy as np
from scipy import ndimage
from scipy import misc

```

```python
mapache = misc.face()
mapache
```

```python
mapache.shape
```

```python
type(mapache)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(mapache)
```

```python
# girar mapache
plt.imshow(mapache[::-1])
```

```python
# duplciar pixeles

plt.imshow(mapache ** 2)
```

```python
# 1 seleccionar cara del mapache

plt.imshow(mapache[0:500, 500:800], cmap=plt.cm.gray)
```

```python
# 2 desenfoque


desenfoque_mapache = ndimage.gaussian_filter(mapache, sigma=8)
plt.imshow(desenfoque_mapache)
```

```python
# 3 Filtro azul sobre la imagen
plt.imshow(mapache[:,:,2],vmin=0,vmax=1620)

```

```python
# 4 filtro rojo
mapache_rojo=np.copy(mapache)
mapache_rojo[:,:,1]=0
mapache_rojo[:,:,2]=0
plt.imshow(mapache_rojo)
```

```python
# flip de la imagen (espejo)

plt.imshow(mapache[0:360])
plt.show()
plt.imshow(mapache[360:720])
plt.show()
```

```python
# 5 rotar mapache 45 y 90 grados
rota45 = ndimage.rotate(mapache, 45)
plt.imshow(rota45)

rota90 = ndimage.rotate(mapache, 90)
plt.imshow(rota90)
```
